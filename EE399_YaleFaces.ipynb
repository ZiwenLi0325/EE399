```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
from numpy import linalg 

results=loadmat('yalefaces.mat')
X=results['X']
print(X.shape)

fig = plt.figure(figsize=(15, 10))
for k in range(9):
    Xp = X[:, k].reshape((32, 32), order='F')
    uplot = Xp.T
    ax = fig.add_subplot(3, 3, k+1)
    ax.imshow(uplot)
```

```python
m=10
Xm=X[:, 0:9]
C=np.matmul(Xm.T,Xm)
print(Xm.shape)
print(C.shape)
print(C)
plt.pcolor(C)
```

```python
u,s,v=np.linalg.svd(X)
fig = plt.figure(figsize=(15, 10))
for k in range(9):
    Up = u[:, k].reshape((32, 32), order='F')
    uplot = Up.T
    ax = fig.add_subplot(3, 3, k+1)
    ax.imshow(uplot)

print(s.shape)
print(s[0:5])
print(v.shape)
print(np.sum(s))
```

```python

plt.stem(s, markerfmt='ro')
plt.xlim(-2, 20)
plt.show()
```
